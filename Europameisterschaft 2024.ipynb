# Europameisterschaft 2024 Vorhersage

```python
import pandas as pd
pd.set_option("display.max_rows", None)
import requests
```

## Aktuelle Wettquoten

```python
url = "https://www.wettbasis.com/em/em-wetten"
r = requests.get(url)
df_list = pd.read_html(r.text) 
df = df_list[1]
df.head(len(df))
```

```python
df=df.rename(columns={"Unnamed: 0": "Länder", "Unnamed: 1": "bet365", "Unnamed: 2":"Oddset","Unnamed: 3":"bwin",
                   "Unnamed: 4":"Betano","Unnamed: 5":"bet-at-home","Unnamed: 6":"Interwetten"})
df.head(len(df))
```

```python
df['Durchschnitt'] = df.select_dtypes(include='number').mean(axis=1).round(2)
df.head(len(df))
```

## Aktuelles FIFA Ranking

```python
url2="https://www.betinf.com/fifa_ranking.htm"
r2 = requests.get(url2)
df_list2 = pd.read_html(r2.text) 
df2 = df_list2[0]
df2.head(len(df2))
```

```python
teams=["England","Frankreich","Deutschland","Spanien","Portugal","Italien","Belgien","Niederlande","Dänemark",
       "Kroatien","Türkei","Schweiz","Serbien","Österreich","Ungarn","Schottland","Ukraine","Polen",
       "Tschechien","Rumänien","Slowenien","Albanien","Slowakei","Georgien"]
```

```python
df2["Team"]=df2["Team"].replace({"France":"Frankreich","Germany":"Deutschland","Spain":"Spanien","Italy":"Italien","Belgium":"Belgien",
                         "Netherlands":"Niederlande","Denmark":"Dänemark","Croatia":"Kroatien","Turkiye":"Türkei",
                         "Switzerland":"Schweiz","Serbia":"Serbien","Austria":"Österreich","Hungary":"Ungarn",
                         "Scotland":"Schottland","Poland":"Polen","Czechia":"Tschechien","Romania":"Rumänien",
                        "Slovenia":"Slowenien","Albania":"Albanien","Slovakia":"Slowakei","Georgia":"Georgien"})
```

```python
df2.head(len(df2))
```

```python
df2 = df2[df2["Team"].isin(teams)].reset_index()
df2 = df2.rename(columns={"Team":"Länder"})
df2.head(len(df2))
```

## Verbinden beider Tabellen

```python
df_join = pd.merge(df, df2, on="Länder", how="inner")
df_join.head(len(df_join))
```

## Erstellen eines Modells
- Annahme Wettquoten sind dynamischer und Aussagekräftiger (Gewichtung 80%)
- FIFA Ranking (Gewichtung 20%)

```python
df_join["Durchschnitt Wahrscheinlichkeit in %"] = round(1/df_join["Durchschnitt"]*100,2)
df_join["Rang Wahrscheinlichkeit in %"] = round(1/df_join["Rank"]*100,2)
df_join.head(len(df_join))
```

```python
gewichtung_wetten = 0.8
gewichtung_fifa_ranking = 0.2
```

```python
#max_rank = df_join["Rank"].max()
#min_rank = df_join["Rank"].min()
#df_join["Normalisierter Rang"] = (max_rank - df_join["Rank"]) / (max_rank - min_rank)
#max_avg = df_join["Durchschnitt"].max()
#min_avg = df_join["Durchschnitt"].min()
#df_join["Normalisierter Durchschnitt"] = (max_avg - df_join["Durchschnitt"]) / (max_avg - min_avg)
```

```python
#df_join["Gewichtetes Ranking"] = ((df_join["Normalisierter Durchschnitt"] * gewichtung_wetten ) + (df_join["Normalisierter Rang"] * gewichtung_fifa_ranking)).round(2)
df_join["Gewichtetes Ranking"] =((df_join["Durchschnitt Wahrscheinlichkeit in %"] * gewichtung_wetten ) + (df_join["Rang Wahrscheinlichkeit in %"] * gewichtung_fifa_ranking)).round(2)
df_final= df_join
df_final = df_final.sort_values(by="Gewichtetes Ranking", ascending=False).reset_index(drop=True)
df_final.head(len(df_final))
```
